```python
import cv2
import numpy as np

video_yolu = "vid_1.avi"
cap = cv2.VideoCapture(video_yolu)

kirmizi_alan = np.array([160, 100, 100])
kirmizi_ust = np.array([180, 255, 255])
beyaz_alan = np.array([0, 0, 200])
beyaz_ust = np.array([180, 25, 255])
sari_alan = np.array([20, 100, 100])
sari_ust = np.array([40, 255, 255])

onceki_kirmizi_merkez = None
onceki_beyaz_merkez = None
onceki_sari_merkez = None
kirmizi_hiz = 0
beyaz_hiz = 0
sari_hiz = 0

kirmizi_top_izi = []
beyaz_top_izi = []
sari_top_izi = []

hamleBas = False

while True:
    ret, kare = cap.read()
    if not ret:
        break

    hsv = cv2.cvtColor(kare, cv2.COLOR_BGR2HSV)

    kirmizi_maskesi = cv2.inRange(hsv, kirmizi_alan, kirmizi_ust)
    beyaz_maskesi = cv2.inRange(hsv, beyaz_alan, beyaz_ust)
    sari_maskesi = cv2.inRange(hsv, sari_alan, sari_ust)

    kirmizi_konturlar, _ = cv2.findContours(kirmizi_maskesi, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    beyaz_konturlar, _ = cv2.findContours(beyaz_maskesi, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    sari_konturlar, _ = cv2.findContours(sari_maskesi, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(kirmizi_konturlar) > 0:
        kirmizi_kontur = max(kirmizi_konturlar, key=cv2.contourArea)
        kirmizi_analiz = cv2.moments(kirmizi_kontur)
    
        if kirmizi_analiz["m00"] > 0:
            kirmizi_merkez = (int(kirmizi_analiz["m10"] / kirmizi_analiz["m00"]),
                              int(kirmizi_analiz["m01"] / kirmizi_analiz["m00"]))
    
            kirmizi_top_izi.append(kirmizi_merkez)
    
            if onceki_kirmizi_merkez is not None:
                hareket_esigi = 2 
                kirmizi_hiz = np.sqrt((kirmizi_merkez[0] - onceki_kirmizi_merkez[0]) ** 2 +
                                      (kirmizi_merkez[1] - onceki_kirmizi_merkez[1]) ** 2)
                if kirmizi_hiz > hareket_esigi:
                    kirmizi_hiz = kirmizi_hiz
                else:
                    kirmizi_hiz = 0
    
            cv2.circle(kare, kirmizi_merkez, 10, (0, 0, 255), -1)
            cv2.putText(kare, f"Kirmizi Hiz: {kirmizi_hiz:.2f} piksel/kare", (10, 30),
                        cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 255), 2)
    
            cv2.rectangle(kare, (kirmizi_merkez[0] - 10, kirmizi_merkez[1] - 10),
              (kirmizi_merkez[0] + 10, kirmizi_merkez[1] + 10), (0, 0, 255), 2)
    
            onceki_kirmizi_merkez = kirmizi_merkez


    if len(beyaz_konturlar) > 0:
        beyaz_kontur = max(beyaz_konturlar, key=cv2.contourArea)
        beyaz_analiz = cv2.moments(beyaz_kontur)
    
        if beyaz_analiz["m00"] > 0:
            beyaz_merkez = (int(beyaz_analiz["m10"] / beyaz_analiz["m00"]),
                            int(beyaz_analiz["m01"] / beyaz_analiz["m00"]))
    
            beyaz_top_izi.append(beyaz_merkez)
    
            if onceki_beyaz_merkez is not None:
                hareket_esigi = 2  
                beyaz_hiz = np.sqrt((beyaz_merkez[0] - onceki_beyaz_merkez[0]) ** 2 +
                                    (beyaz_merkez[1] - onceki_beyaz_merkez[1]) ** 2)
                if beyaz_hiz > hareket_esigi:
                    beyaz_hiz = beyaz_hiz
                else:
                    beyaz_hiz = 0
    
            cv2.circle(kare, beyaz_merkez, 10, (255, 255, 255), -1)
            cv2.putText(kare, f"Beyaz Hiz: {beyaz_hiz:.2f} piksel/kare", (10, 60),
                        cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 255, 255), 2)
    
            cv2.rectangle(kare, (beyaz_merkez[0] - 10, beyaz_merkez[1] - 10),
              (beyaz_merkez[0] + 10, beyaz_merkez[1] + 10), (255, 255, 255), 2)

    
            onceki_beyaz_merkez = beyaz_merkez
    
    if len(sari_konturlar) > 0:
        sari_kontur = max(sari_konturlar, key=cv2.contourArea)
        sari_analiz = cv2.moments(sari_kontur)
    
        if sari_analiz["m00"] > 0:
            sari_merkez = (int(sari_analiz["m10"] / sari_analiz["m00"]),
                           int(sari_analiz["m01"] / sari_analiz["m00"]))
    
            sari_top_izi.append(sari_merkez)
    
            if onceki_sari_merkez is not None:
                hareket_esigi = 2 
                sari_hiz = np.sqrt((sari_merkez[0] - onceki_sari_merkez[0]) ** 2 +
                                   (sari_merkez[1] - onceki_sari_merkez[1]) ** 2)
                if sari_hiz > hareket_esigi:
                    sari_hiz = sari_hiz
                else:
                    sari_hiz = 0
    
            cv2.circle(kare, sari_merkez, 10, (0, 255, 255), -1)
            cv2.putText(kare, f"Sari Hiz: {sari_hiz:.2f} piksel/kare", (10, 90),
                        cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 255, 255), 2)
    
            cv2.rectangle(kare, (sari_merkez[0] - 10, sari_merkez[1] - 10),
              (sari_merkez[0] + 10, sari_merkez[1] + 10), (0, 255, 255), 2)
    
            onceki_sari_merkez = sari_merkez

    for i in range(1, len(kirmizi_top_izi)):
        cv2.line(kare, kirmizi_top_izi[i - 1], kirmizi_top_izi[i], (0, 255, 0), 2)
        cv2.circle(kare, kirmizi_top_izi[0], 5, (0, 255, 0), -1)
        cv2.circle(kare, kirmizi_top_izi[-1], 5, (0, 255, 0), -1)
        cv2.putText(kare, "Baslangic", kirmizi_top_izi[0], cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        cv2.putText(kare, "Bitis", kirmizi_top_izi[-1], cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    if beyaz_hiz >2 and len(beyaz_top_izi) != 0:
        hamleBas=True
        cv2.putText(kare, "Hamle Basladi", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        
    if hamleBas==True and beyaz_hiz <1 and len(beyaz_top_izi) > 0 and kirmizi_hiz <1 and len(kirmizi_top_izi) > 0 and sari_hiz<1 and len(sari_top_izi) > 0:
        cv2.putText(kare, "Hamle Bitti", (10, 150), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow("Video", kare)

    if cv2.waitKey(25) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
